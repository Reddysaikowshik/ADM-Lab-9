```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("/content/salary_data.csv")
```

```python
df.head()
```

```python
plt.scatter(df["YearsExperience"], df["Salary"], color='red')
plt.title("Years of Experience vs Salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
X = df[["YearsExperience"]].values
y = df["Salary"].values
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly,y)
```

```python
y_pred = model.predict(X_poly)
```

```python
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred) # Now r2_score is defined and can be used
print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2:.4f}")
```

```python
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = poly.transform(X_plot)
y_plot = model.predict(X_plot_poly)
```

```python
plt.scatter(X, y, color='red', label='Actual Data')
plt.plot(X_plot, y_plot, color='blue', label='Polynomial Regression Curve')
plt.title("Polynomial Regression: Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```
